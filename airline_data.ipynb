```python
# Revisamos la frecuencia de cada valor de la variable EDUCATION
df_loyalty['Education'].value_counts()
```

```python
# Seleccionamos todas las variables categóricas y lo añadimos en una nueva variable para poder trabajar mejor
df_loyalty_cat = df_loyalty.select_dtypes(include='object')
df_loyalty_cat.head(2)
```

```python
# Eliminamos las columnas que no son de nuestro interés usando el método '.drop()', en este caso la columna del CP
df_loyalty_cat_drop = df_loyalty_cat.drop('Postal Code', axis=1, inplace=False)
df_loyalty_cat_drop
```

```python
# Para utilizar unique() y value_counts() haremos un bucle for, pero antes creamos una variable con las columnas del DF
columnas_cat = df_loyalty_cat_drop.columns
print(f"Las columnas del DataFrame de variables categóricas son {columnas_cat}")
```

```python
# Con el bucle for, iteramos por cada una de las columnas para sacar sus valores únicos y sus frecuencias

for columna in columnas_cat:
    resultado_u = df_loyalty_cat_drop[columna].unique()
    resultado_c = df_loyalty_cat_drop[columna].value_counts()
    print(f"Columna: {columna}")
    print(f"Valores únicos: {resultado_u}")
    print(f"Frecuencias de dichos valores: \n{resultado_c}\n")
```
